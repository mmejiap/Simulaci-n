## Problema 01
Utilice el m√©todo de la transformada inversa para generar valores de una Variable Aleatoria X en el que
$ P[X=i] = \frac{e^{-\lambda} \lambda^i / i!}{\sum_{j=0}^{k} e^{-\lambda} /j!}\space\space\space\space i=0,1,...,k$
#### Sol

```r
rm(list=ls())

GeneraPoissontruncada <- function(lambda, k, n){
    X<-rep(0,n)
    for(j in 1:n){
        i <- 0
        p <- exp(-lambda)/ppois(k, lambda)
        F <- p
        u<-runif(1,0,1)
        while(u>=F){
            p<-(lambda*p)/(i+1)
            F<-F+p
            i<-i+1
        }
        X[j]<-i
    }
    return(X)
}


k<-20
lambda<-2
n<-1000

X <- GeneraPoissontruncada(lambda,k,n)

xx <- table(X)/n

barplot(xx,main = "Poisson Truncada - Transformada Inversa")
```

## Problema 02
Sea $X\neg Bin(n,p)$. Suponga que queremos generar una Variable Aleatoria $Y$ cuya funci√≥n de masa de probabilidad es igual a la funci√≥n de masa de probabilidad de $X$ dado que $X\geq k$ para alg√∫n $k\leq n$. Sea $\alpha=[X \geq k]$ y suponga que se ha calculado el valor de $\alpha$ . Utilice el m√©todo de la transformada inversa para generar valores de $Y$
#### Sol

```r
rm(list=ls())
GeneralBinomialTruncada <- function(n, pp, k, nn){
    x<-rep(0,nn)
    for(j in 1:nn){
        i<-k
        p<-dbinom(i,n,pp)/(1-pbinom(k-1,n,pp))
        F<-p
        u<-runif(1,0,1)
        while(u>=F){
            p<-((n-i)/(i+1))*(pp/(1-pp))*p
            F<-F+p
            i<-i+1
        }
        x[j]<-i
    }
    return(x)
}

k<-3
n<-10
pp<-0.6
nn<-1000

x<-GeneralBinomialTruncada(n,pp,k,nn)
xx<-table(x)/nn
barplot(xx,main="Binomial Truncada - Transformada Inversa")


```

## Problema 03
Emplear 1000 numeros aleatorios para explicar la forma de aproximar $\sum^{N}_{i=1} e^{i/N}$ con $N=1000$

$$ \sum^{N}_{i=1} \frac{a(i)}{N} = \sum^{N}_{i=1} e^{i/N}$$
#### Sol

- $ a(i) = Ne^{i/N} $

```r
rm(list=ls())

N<-1000
NN<-1:N

k<-100
kk<-1:k

i<-floor(N*runif(k,0,1))

a_i <- N*exp(i/N)

mediaAcum <- cumsum(a_i)/kk

plot(kk,mediaAcum,type='l')

mediaAcum[length(kk)]

sum(exp(NN/N))
```

## Problema 04

Se lanza de manera continua un par de dados legales, hasta que todos los posibles resultados $2,3,‚Ä¶,12$ hayan aparecido al menos una vez. Desarrolle un estudio de simulaci√≥n para estimar el n√∫mero esperado de lanzamientos necesarios
#### Sol


```r
rm(list=ls())

s<-1000
X<-rep(0,s)

for(i in 1:s){
    suma<-2:12
    j<-0
    suma1<-rep(1,length(suma))
    x<-sum(floor(6*runif(2,0,1))+1)
    while(sum(suma1)>0){
        r<-1
        while(suma[r]!=x && r<=length(suma)){
            r<-r+1
        }
        suma1[r]<-0
        j<-j+1
        x<-sum(floor(6*runif(2,0,1))+1)
    }
    X[i]<-j
}

ii<-1:s
XX<-cumsum(X)/ii
plot(ii,XX,type='l')
XX[length(XX)]
```
