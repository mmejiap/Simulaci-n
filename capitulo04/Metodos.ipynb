## Generacion de variables aleatorias de Poisson - discretas - metodo transformada inversa

```r
genPoisson_discreto_ti <- function(k,lambda){
    X<-rep(0,k)
    n<-1:k
    for(j in n){
        u<-runif(1,0,1)
        F<-0
        i<-0
        #p<-exp(-lambda)*(lambda^i)/factorial(i)
        p<-exp(-lambda)
        while(u>=F){
            p<-lambda*p/(i+1)
            F<-F+p
            i<-i+1
        }
        X[j]<-i
    }
    return(X)
}
X <- genPoisson_discreto_ti(1000,20)
hist(X)
```

## Generacion de variables aleatorias de Binomial - discretas - metodo transformada inversa

$P[X=i] = \frac{n!}{i!(n-2)!} ^i(1-p)^{n-i}$ ,  $ i = 0,1,...,n$ 

```r
genBinomial_discreto_ti <- function(k,p){
    n<-1:k
    X<-rep(0,k)
    for(j in n){
        i <- 0
        c <- p/(1-p)
        pr<- (1-p)^k
        F<-pr
        u<-runif(1,0,1)
        while(u>=F){
            pr<-(k-i)*c*pr/(i+1)
            F<-F+pr
            i<-i+1
            
        }
        X[j]<-i
    }
    return(X)
}

X<-genBinomial_discreto_ti(1000,0.5)
hist(X)
```

```r
vectorBinomial <- function(n,p){
    X<-rep(0,n)
    for(j in 1:n){
        c<-p/(1-p)
        i<-0
        pr <- (1-p)^n
        F<-pr
        u<-runif(1,0,1)
        while(u>=F){
            pr<-(n-i)*c*pr/(i+1)
            F<-F+pr
            i<-i+1
            
        }
        X[j]<-i
    }
    return(X)
}

xx<-vectorBinomial(100,0.5)
hist(xx)
```

```r
poisson <- function(n, lambda){
    X<-rep(0,n)
    for(j in 1:n){
        p <- exp(-lambda)
        i<-0
        F<-p
        u <- runif(1,0,1)
        while(u>=F){
            p<-lambda*p/(i+1)
            F<-F+p
            i<-i+1
            
        }
        X[j]<-i
    }
    return(X)
}

hist(poisson(1000,5))
```
