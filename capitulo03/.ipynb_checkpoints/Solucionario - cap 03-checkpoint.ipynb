### Problema 01
Si $x_{0}=5$ y $x_{n} = 3x_{n-1}\space mod \space 150$
    determine $x_1, ... , x_{10}$
    
    
#### Sol.


```r
rm(list = ls())
k<-10
x <- rep(0,k)
x[1] <- 5
for(i in 2:(k+1)){
     x[i] <- (3*x[i-1])%%150
}
print(x[2:(k+1)])
```

### Problema 03
$ \int_{1}^{0} \! exp[e^x] \, dx$

#### Sol.

```r
rm(list=ls())

k<-500
n<-1:k

u=runif(k,0,1)

int_aprox <- cumsum(exp(exp(u)))/n

plot(int_aprox, type = 'l')

#valor aproximado
print(int_aprox[length(u)])

#valor real en R
f<-function(x) exp(exp(x))
print(integrate(f,0,1))
```

### Problema 04
$\int_{0}^{1} \! (1-x^2)^{\frac{3}{2}} \, dx$

#### Sol.

```r
rm(list = ls())

k<-500
n<-1:k

u <- runif(k,0,1)

int_aprox <- cumsum((1-u^2)^(3/2))/n

plot(int_aprox, type='l',xlab = 'n',ylab = 'valor', main = 'Pregunta 04')

#valor aproximado
print(int_aprox[length(u)])


#valor real en R
f<-function(x) (1-x*x)^(3/2)
print(integrate(f,0,1))
```

### Problema 06
$\int_{0}^{\infty} \! x(1+x^2)^{-2} \, dx$
#### Sol.

```r
rm(list = ls())

k<-1000
n<-1:k

u<-runif(k,0,1)

f <- function(x) (x*(1+x^2)^(-2))

int_aprox <- cumsum(f(1/u-1)/(u^2))/n
plot(int_aprox,type = 'l',xlab = 'n',ylab = 'valor', main = 'Pregunta 06')

#valor aproximado
print(int_aprox[length(u)])

#valor calculado
print(integrate(f,0,Inf))
```

### Problema 07
$\int_{-\infty}^{\infty} \! e^{-x^2} \, dx$
#### Sol.

Dado que $e^{-x^2}$ es una funcion par, entonces la integral anterior quedaria $2 \int_{0}^{\infty} \! e^{-x^2} \, dx$

```r
rm(list = ls())

k<-500
n<-1:k

u=runif(k,0,1)

f <- function(x) exp(-x^2)

int_aprox <- cumsum( 2*f(1/u-1)/(u^2) )/n

#valor aproximado
print(int_aprox[length(u)])
plot(int_aprox,type = 'l',xlab = 'n',ylab = 'valor', main = 'Pregunta 07')

#valor calculado
print(integrate(f,-Inf,Inf))
    
```

### Problema 08
$\int_{0}^{1} \int_{0}^{1} \! e^{(x+y)^2} \, dy \, dx$
#### Sol.

```r
rm(list = ls())

k<-1000
n<-1:k

x<-runif(k,0,1)
y<-runif(k,0,1)

f<-function(x,y) exp((x+y)^2)
                            
int_aprox <- cumsum( f(x,y) )/n

#valor aproximado
print(int_aprox[length(x)])
plot(int_aprox,type = 'l',xlab = 'n',ylab = 'valor', main = 'Pregunta 08')

#valor calculado
#print(integrate2(f, 0, 1, 0, 1))
#library('rmutil')
#help(int2)
print(int2(f, a=c(0,0), b=c(1,1)))
```

### Problema 10
Utilice la simulacion para aproximar $ Cov(U,e^U)$, donde $U$ es uniforme en $(0,1)$.\
Compare su aproximacion con la respuesta correcta
#### Sol.

```r
rm(list = ls())

k <- 1000
n <- 1:k

u <- runif(k,0,1)

cov = cumsum(u*exp(u))/n - (cumsum(u)/n )*(cumsum(exp(u))/n)

plot(cov,type='l',xlab='n',ylab = 'valor',main='Problema 10')

print(paste("Covarianza -> ",cov[length(u)]))
```

a) $Cov(U,\sqrt{1-U^2})$

#### Sol.

```r
#Prob 11a

rm(list=ls())

k <- 1000
n <- 1:k

u <- runif(k,0,1)

cov <- cumsum(u*sqrt(1-u^2))/n - ( cumsum(u)/n )*( cumsum(sqrt(1-u^2))/n )

plot(cov,type='l',xlab='n',ylab = 'valor',main='Problema 11-a')

print(paste("Covarianza -> ",cov[length(u)]))
```

### Problema 12
Para $U_1, U_2, ... $ variables aleatorias uiformes en $(0,1)$, definimos

$ N = Minimo \left\{
	       \begin{array}{ll}
		 n:&\displaystyle\sum_{i=1}^{n} U_i  > 1 \\
		 \end{array}
 	     \right\}
$

Es decir, $N$ es igual a la cantidad de números aleatorios que deben sumarse para exceder a 1

a) Estime $E[N]$ generando 100 valores de $N$

b) Estime $E[N]$ generando 1000 valores de $N$

c) Estime $E[N]$ generando 10000 valores de $N$

d) ¿Cuál cree que sea el valor de $E[N]$?

#### Sol.

```r
# Problema 12
rm(list=ls())

k<-10000
N=rep(0,k)

for(i in 1:k){
    u<-runif(1,0,1)
    suma<-0
    n<-0
    while(suma<1){
        suma<-suma+u
        u<-runif(1,0,1)
        n<-n+1
    }
 N[i]<-n   
}
esperanza=cumsum(N)/(1:k)
plot(esperanza,type='l',xlab='n',ylab='valor',main='Problema 12')

print(paste("Esperanza -> ",esperanza[k]))
```

### Problema 13
Para $U_1, U_2, ... $ variables aleatorias uiformes en $(0,1)$, definimos

$ N = Máximo \left\{
	       \begin{array}{ll}
           n: &\displaystyle\prod_{i=1}^{n} U_{i} \geq e^{-3} \\
		 \end{array}
 	     \right\}
$

Donde $\prod_{i=1}^{0} U_i \equiv 1$.

a) Determine $E[N]$ por simulacion.

b) Determine $P[N=i]$, para $i=0,1,2,3,4,5,6$, por simulacion


#### Sol.

```r
# Problema 13

k <- 500

N <- rep(0,k)

for(i in 1:k){
    prod<-1
    u<-runif(1,0,1)
    n<-0
    while(prod>=exp(-3)){
        prod<-prod*u
        u<-runif(1,0,1)
        n<-n+1
    }
    N[i]<-n
}

esperanza <- cumsum(N)/(1:k)

plot(esperanza,type='l')

# parte a)
print(paste("E[N] = ", esperanza[k]))

# parte b)
print((table(N)/k)[1:7])
```
